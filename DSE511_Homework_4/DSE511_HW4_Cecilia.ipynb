## Part A — Sorting Showdown

**Goal:** Empirically compare sorting algorithms and relate runtime to big‑O.

You will:

2. Sort using:
   - Python built‑in `sorted()` (Time to sort ~ O(n log n) average/worst on random data)
   - Your own **insertion sort** (O(n²))
   - One additional algorithm of your choice (e.g., `numpy.sort`); must add

```python
#Implement insertion sort for any argument "arr"

#1. I used this website to help me code it: https://www.geeksforgeeks.org/python/python-program-for-insertion-sort/
def insertionSort(arr):
    a = list(arr) 
    n = len(a)
    if n <= 1:
        return a
    for i in range(1, n):
        key = a[i]
        j = i - 1
        while j >= 0 and key < a[j]:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
    return a  # returns sorted a copy of arr

# 2. To check if the function ran correcrtly on a sample, this will read "Insertion sort sanity check passed."
sample = [5, 2, 4, 6, 1, 3]
assert insertionSort(sample) == sorted(sample)
print("Insertion sort sanity check passed.")
```

```python
#3. Now I will generate integer arrays of different sizes
# - 4 different array sizes repeated 3 times
sizes = [10**3, 10**4, 10**5, 10**6] 
repeat = 3

#4. Cap insertion sort at manageable n to prevent run from taking too long
ins_cap = 20_000
```

```python
#I'm interested in comparing sort() and insertionSort to quicksort, since it's considered one of the best sorting methods
#5. Use numpy_sort function since it's python's quicksort

#6. First, intialize the results container to collect the timing results with the following columns
results = {
    'n': [], # Number of elements in the array
    'sorted_builtin': [], # Timing results for Python's built-in sorted()
    'insertion_sort': [], # Timing results for custom insertion sort
    'quicksort': [] # Timing results for NumPy's np.sort (default quicksort)
}

#7. Loop over array sizes to benchmark
for n in sizes:
    # Generate random array of n size where
    # - low=0, high=10_000_000 sets the range of values
    # - dtype=np.int64 ensures consistent integer type across runs
    data = rng.integers(low=0, high=10_000_000, size=n, dtype=np.int64)

    # Time Python built-in sorted()
    t_sorted, _ = time_function(sorted, data, repeat=repeat)
    
    # Time insertion sort (only if n <= ins_cap)
    if n <= ins_cap:
        t_insert, _ = time_function(insertionSort, data, repeat=repeat)
    else:
        t_insert = np.nan

    # Time NumPy sort
    t_numpy, _ = time_function(np.sort, data, repeat=repeat)
    
    # Append results
    results['n'].append(n)
    results['sorted_builtin'].append(t_sorted)
    results['insertion_sort'].append(t_insert)
    results['quicksort'].append(t_numpy)

#8. Convert results dictionary to a pandas DataFrame
import pandas as pd
dfA = pd.DataFrame(results)
# Display the benchmarking results
dfA
```

## Part B — Data Structures in Action: Membership Testing

**Goal:** Compare membership query performance for `list` (≈ O(n)), `set` (≈ O(1) average), and `dict` (≈ O(1) average).

You will:
1. Build collections of increasing size of unique IDs/strings.
2. Generate random queries that are both present and absent.
3. Measure membership test times (`x in structure`) for each structure type.
4. Plot how query time scales with `n`.


```python
#1. Create a set of 3 sample sizes and generate 1,000 random queries per sample size
sizes_B = [10_000, 100_000, 1_000_000]
queries_per_size = 1_000

#2. Generate random queries by generating "n" number of ids and randomizes their order
def gen_ids(n):
    ids = [f"id_{i:07d}" for i in range(n)]
    rng.shuffle(ids)
    return ids

#3. Create a memership timing function which simulates both successful and unsuccessful lookups for each type of data structure
def membership_timing(ids, structure_type='list'):
    # Construct collection depending on type
    if structure_type == 'list':
        coll = list(ids)
        def contains(x): return x in coll
    elif structure_type == 'set':
        coll = set(ids)
        def contains(x): return x in coll
    elif structure_type == 'dict':
        coll = {x: True for x in ids}
        def contains(x): return x in coll

    #build queries
    present = rng.choice(ids, size=queries_per_size//2, replace=False) # IDs that exist
    absent = [f"nope_{i:07d}" for i in range(queries_per_size - len(present))] # IDs that don't exist
    queries = np.concatenate([present, absent])
    rng.shuffle(queries)

    #Time the membership queries
    start = time.perf_counter()
    for q in queries:
        _ = contains(q)
    elapsed = time.perf_counter() - start
    return elapsed

#4. Run benchmark and collect results
records = {'n': [], 'structure': [], 'time_s': []}

for n in sizes_B:
    ids = gen_ids(n)
    for stype in ['list', 'set', 'dict']:
        t = membership_timing(ids, structure_type=stype)
        records['n'].append(n)
        records['structure'].append(stype)
        records['time_s'].append(t)

dfB = pd.DataFrame(records)

# Pivot to wide format
dfB_wide = dfB.pivot(index='n', columns='structure', values='time_s').reset_index()
dfB_wide = dfB_wide[['n', 'list', 'set', 'dict']]

dfB_wide
```

```python
#Final step: Plot the results

#1. generate a new figure
plt.figure()

#2. loop over each data structure type
for stype in ['list', 'set', 'dict']:
    #select a subset of results for the current structure
    sub = dfB[dfB['structure'] == stype]
    # Plot n (dataset size) vs time_s (elapsed membership query time)
    # - marker='o' puts a circle at each data point for clarity
    # - label=stype ensures it shows up in the legend
    plt.plot(sub['n'], sub['time_s'], marker='o', label=stype)

#3. add axis labels
plt.xlabel('n')
plt.ylabel('Total time for membership queries (s)')

#4. add a title
plt.title('Membership testing time vs n')

#5. add a legend
plt.legend()

#6. show the plot
plt.show()
```
